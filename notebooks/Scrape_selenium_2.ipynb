```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
```

```python
URL = "https://twitter.com/DeepMind/status/1494041801572990979"
CHROME_PATH = "/home/harshapk/bin/chromedriver"
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-setuid-sandbox')
s=Service(ChromeDriverManager().install())
browser = webdriver.Chrome('chromedriver', options=chrome_options) # Doesn't work in colab for some reason
```

```python
browser.maximize_window()
browser.get(URL)
timeline_texts = []
body = browser.find_element_by_tag_name('body')
try:
    timeline_texts.append(body.find_element_by_css_selector("[aria-label='Timeline: Conversation']").text)
except:
    pass
ct=0
browser.get_screenshot_as_file(f"capture{ct}.png")
ct+=1
for _ in range(50):
    body.send_keys(Keys.PAGE_DOWN)
    #browser.get_screenshot_as_file(f"capture{ct}.png")
    try:
        timeline_texts.append(body.find_element_by_css_selector("[aria-label='Timeline: Conversation']").text)
    except:
        pass
    ct+=1
browser.get_screenshot_as_file(f"capture{ct}.png")
timeline_texts.append(body.find_element_by_css_selector("[aria-label='Timeline: Conversation']").text)
timeline = body.find_element_by_css_selector("[aria-label='Timeline: Conversation']")
ct+=1
html = browser.page_source
body = browser.find_element_by_tag_name('body')
```
